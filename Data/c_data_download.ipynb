```stata
*Pulling EF zip file from IPEDS
copy "https://nces.ed.gov/ipeds/datacenter/data/EF2022A_Data_Stata.zip" EF2022A_Data_Stata.zip, replace
```

```stata
*Unzipping the EF file
unzipfile EF2022A_Data_Stata.zip
```

```stata
*Pulling HD zip file from IPEDS
copy https://nces.ed.gov/ipeds/datacenter/data/HD2022_Data_Stata.zip HD2022_Data_Stata.zip, replace
```

```stata
*Unzipping the HD file
unzipfile HD2022_Data_Stata.zip
```

```stata
*Loading in the EF.csv file
insheet using ef2022a_data_stata.csv, clear
```

```stata
*Keeping all efalevel 1 observations
keep if efalevel==1
```

```stata
save ef2022.dta, replace
```

```stata
*Loading in the HD.csv file
insheet using hd2022_data_stata.csv, clear
```

```stata
*All the codes up to this stage were for preparing the two datasets for merging.
*The 1s mean the unitids in each of the datasets are being merged.
merge 1:1 unitid using ef2022.dta
```

```stata
*In the result section, the number of observations mentioned in front of "from using" is 0, which means the code worked.
```

```stata
%head
```

```stata
keep unitid eftotlt eftotlm eftotlw efhispt efhispm efhispw efnralt efnralm efnralw instnm obereg sector ugoffer groffer
```

```stata
*Labelling all the selected variables
label variable unitid "School Unique Identification Number"
label variable eftotlt "Grand Total"
label variable eftotlm "Grand Total Men"
label variable eftotlw "Grand Total Women"
label variable instnm "Institution Name"
label variable obereg "BEA Regions"
label variable sector "Sector of Institution"
label variable ugoffer "Undergraduate Offering"
label variable groffer "Graduate Offering"
label variable efhispt "Hispanic Total"
label variable efhispm "Hispanic Men"
label variable efhispw "Hispanic Women"
label variable efnralt "US Non-Resident Total"
label variable efnralm "US Non-Resident Men"
label variable efnralw "US Non-Resident Women"
```

```stata
*Creating dummy variables fro the region categories and labelling them:
gen newengland=obereg==1
label var newengland "New England"

gen mideast=obereg==2
label var mideast "Mid East"

gen greatlakes=obereg==3
label var greatlakes "Great Lakes"

gen plains=obereg==4
label var plains "Plains"

gen southeast=obereg==5
label var southeast "Southeast"

gen southwest=obereg==6
label var southwest "Southwest"

gen rockymountains=obereg==7
label var rockymountains "Rocky Mountains"

gen farwest=obereg==8
label var farwest "Far West"

gen otherusjurisdictions=obereg==9
label var otherusjurisdictions "Other US Jurisdictions"
```

```stata
*Making the demographic variables into percentiles

gen pctefhispt=efhispt / eftotlt * 100
gen pctefhispm=efhispm / eftotlm * 100
gen pctefhispw=efhispw / eftotlw * 100
gen pctefnralt=efnralt / eftotlt * 100
gen pctefnralm=efnralm / eftotlm * 100
gen pctefnralw=efnralw / eftotlw * 100
```

```stata
*Relabelling the newly generated variables 
label variable pctefhispt "Hispanic Total Percent"
label variable pctefhispm "Hispanic Men Percent"
label variable pctefhispw "Hispanic Women Percent"
label variable pctefnralt "US Non-Resident Total Percent"
label variable pctefnralm "US Non-Resident Men Percent"
label variable pctefnralw "US Non-Resident Women Percent"
```
